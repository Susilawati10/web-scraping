```python
def clean_name(name_str):
    # Menghapus karakter yang tidak diperlukan dan spasi ekstra
    clean_str = re.sub(r'[^\w\s]', '', name_str).strip()
    return clean_str

for item in data_name:
    item['name'] = clean_name(item['name'])

print(data_name)
```

```python
import re

def clean_rating(rating_str):
    # Menghapus simbol-simbol yang tidak diperlukan
    clean_str = re.sub(r'[^\d.]', '', rating_str)
    return float(clean_str)

for item in data_rating:
    item['rating'] = clean_rating(item['rating'])

print(data_rating)
```

**Langkah:**

-Setelah pembersihan, string yang dihasilkan diubah menjadi tipe data float.

**Dampak pada Integritas Data:**

-Dengan menghapus simbol yang tidak diperlukan, data rating menjadi valid dan dapat dihitung dengan tepat.

```python
from sqlalchemy import create_engine
import pandas as pd

def read_sqlite(database, table_name):
    conn_string = f'sqlite:///{database}'
    conn = create_engine(conn_string)

    with conn.connect() as db:
        query = f'SELECT * FROM {table_name}'
        df = pd.read_sql(query, con=db)

    return df

# Membaca data dari SQLite
sqlite_database = 'database_name.db'
sqlite_table = 'Apart'

df_sqlite = read_sqlite(sqlite_database, sqlite_table)
print(df_sqlite)
```

### 5. Setelah data disimpan, lakukan query sederhana untuk menarik data dari PostgreSQL dan tampilkan beberapa contoh hasil query tersebut. Jelaskan bagaimana data ini dapat digunakan untuk analisis lebih lanjut atau aplikasi lainnya.

```python
# Menarik Semua Data
def get_all_data(database, table_name):
    conn_string = f'sqlite:///{database}'
    conn = create_engine(conn_string)

    with conn.connect() as db:
        query = f'SELECT * FROM {table_name}'
        df = pd.read_sql(query, con=db)

    return df

# Menarik semua data
all_data = get_all_data(sqlite_database, sqlite_table)
print(all_data.head())  # Menampilkan 5 baris pertama
```

```python
#Menghitung Rata-Rata Rating
def get_average_rating(database, table_name):
    conn_string = f'sqlite:///{database}'
    conn = create_engine(conn_string)

    with conn.connect() as db:
        query = f'SELECT AVG(rating) as average_rating FROM {table_name}'
        avg_rating = pd.read_sql(query, con=db)

    return avg_rating

# Menghitung rata-rata rating
average_rating = get_average_rating(sqlite_database, sqlite_table)
print(average_rating)
```
